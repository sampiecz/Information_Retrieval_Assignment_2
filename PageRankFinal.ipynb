# Import routes.csv, run page rank, list top 15 airports. 

```
'''
Get data and import page rank
'''
import networkx as nx
import csv
import matplotlib.pyplot as plt

def pageRank(alpha, filename, howmany):
  pages = []
  outlinks = []
  with open(filename, newline="") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    for count, row in enumerate(reader):
      pages.append(row[0])
      outlinks.append(row[1])

  G = nx.DiGraph()
  edges = [(page, outlink) for page, outlink in zip(pages, outlinks)]
  G.add_edges_from(edges)
  pr = nx.pagerank(G, alpha)

  print("=========================================")
  print("Top {} - Alpha {}".format(howmany, alpha))
  print("=========================================")

  ranks = []
  names = []

  for count, item in enumerate(sorted(pr.items(), reverse=True)):
    
    if count == howmany:
      break
    else:
      ranks.append(item[1])
      names.append(item[0])
      print( str(count + 1)+ ":", item[0], "-", item[1])

  plt.bar(names, ranks, color="salmon")
  plt.title("{} - {}".format(filename, alpha))
  plt.xticks(rotation=90)
  plt.show()
  
```

```
pageRank(.95, "routes.csv", 15)
```

```
pageRank(.85, "routes.csv", 15)
```

```
pageRank(.75, "routes.csv", 15)
```

The top airports / alphas are relatively similiar. Based on the above bar charts
